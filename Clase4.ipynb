```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head(2)
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )
plt.title('Scatter plot of Wingspan vs CWDistance', fontsize = 17)
plt.xticks(fontsize = 15)
plt.show()
```

```python
sns.swarmplot(x = 'Gender',
              y = 'Wingspan',
              data = df
             )
```

```python
sns.boxplot(data = df.loc[:, ['Age', 'Height', 'Wingspan', 'CWDistance']])
```

```python
sns.boxplot(data = df['Age'])
```

```python
sns.boxplot(data = df.loc[df['Gender'] == 'F',
                      ['Age', 'Height', 'Wingspan', 'CWDistance']],    
           ).set_title('Box plot de genero F')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1) #1 columna, 2 renglones, posicion 1
sns.boxplot(data = [df.loc[df['Gender'] == 'F',
                      'Wingspan']]).set_title('Box plot de Wingspan de genero F')


ax2 = fig.add_subplot(1, 2, 2) #1 columna, 2 renglones, posicion 2
sns.boxplot(data = [df.loc[df['Gender'] == 'M',
                      'Wingspan']]).set_title('Box plot de Wingspan de genero M')
```

```python
sns.distplot(df.CWDistance,
             kde = False) #Se puede quitar este para q aparezca una linea
```

```python
sns.countplot(x = 'Gender', data = df)
```

```python
df = pd.read_csv('datasets/tips.csv')
```

```python
sns.distplot(df['total_bill'],
            kde = False,
            bins = 20).set_title('Histogram of total bill')
```

```python
sns.boxplot(x = df['total_bill'],
           whis = (0,100), #Esto aumenta el rango, si lo quitas aparecen los valores atipicos
            y = df['day']
           ).set_title('Box plot of total bill')
```

```python
fig, ((ax1, ax2, ax3),
      (ax4, ax5, ax6),
      (ax7, ax8, ax9)
     ) = plt.subplots(3, 3, sharex = True, sharey = True)

ax1.plot(2, 3, 'x')
```
