```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head(2)
```

```python
df.columns
```

```python
sns.boxplot(data = df['cintura'])
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'edad',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'median',
                          columns = 'sexo',
                          margins = True
                         )
my_pivot
```

```python
sns.boxplot(data = my_pivot[1])
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'cadera',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'median',
                          columns = 'sexo',
                          margins = True
                         )
my_pivot
```

```python
my_pivot[2].plot()
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'sistol',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'sum',
                          columns = 'sexo',
                          margins = True
                         )
my_pivot
```

```python
my_pivot = my_pivot.drop('All', axis = 0)
```

```python
my_pivot[1].plot(kind = 'bar', figsize = (20,10))
```
