```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
```

```python
%ls
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df.head(2)
```

```python
df.columns
```

```python
modelA = sm.OLS.from_formula('peso~cadera', data = df)
resultA = modelA.fit()
resultA.summary()
```

```python
y = resultA.fittedvalues
y.median()
```

```python
df.corr()
```

```python
modelB = sm.OLS.from_formula('diastol~sistol', data = df)
resultB = modelB.fit()
resultB.summary()
```

```python
modelC = sm.OLS.from_formula('cadera~peso', data = df)
resultC = modelC.fit()
resultC.summary()
```

```python
modelD = sm.OLS.from_formula('diastol~sistol+edad+peso+cintura', data = df)
resultD = modelD.fit()
resultD.summary()
```
