```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df.columns
```

```python
vars = ['SMQ020', 'RIAGENDR', 'RIDAGEYR', 'BMXWT', 'BMXHT', 'BMXBMI',
       'BMXWAIST', 'HIQ210']
```

```python
df = df[vars].dropna()
df
```

```python
df.corr()
```

```python
# BMXWT = b0 + b1*RIDAGEYR
#Modelo base
```

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR', data = df)
result = model.fit()
result.summary()
```

```python
# BMXWT = b0 + b1*RIDAGEYR + b2*RiaGENDR
# BMXWTN = b0 + b1*RIDAGEYR + b2(Male + Female)
#Modelo base
```

```python
df['RiaGENDR'] = df.RIAGENDR.replace({1:'Male', 2:'Female'})
df
```

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR+RiaGENDR', data = df)
result = model.fit()
result.summary()
```
