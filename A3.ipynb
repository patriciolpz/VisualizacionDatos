### A3: Pivot tables

```python
import pandas as pd
import numpy as np
```

```python
%ls
```

```python
df = pd.read_csv('antropometria-dataset-2.csv')
df.head(2)
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'peso',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'mean',
                          columns = 'sexo',
                          margins = True
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'peso',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'median',
                          columns = 'sexo',
                          margins = True
                         )
my_pivot
```

```python
my_pivot.loc['All'].min()
```

```python
df = pd.read_csv('adult.csv')
df.head(2)
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'marital-status',
                          values = 'hours-per-week',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'mean',
                          columns = 'education',
                          margins = True
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'marital-status',
                          values = 'age',
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'median',
                          columns = ['education','sex'],
                          margins = True
                         )
my_pivot
```
