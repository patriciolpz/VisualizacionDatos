```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
%ls
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head(2)
```

```python
df.columns
```

## Histogramas de 'Sistol' y 'Diastol'

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1) #1 columna, 2 renglones, posicion 1
df['sistol'].plot(kind = 'hist', figsize = (20,10), bins = 15)
plt.title('Histograma de Sistol', fontsize = 15)


ax2 = fig.add_subplot(1, 2, 2) #1 columna, 2 renglones, posicion 2
df['diastol'].plot(kind = 'hist', figsize = (20,10))
plt.title('Histograma de Diastol', fontsize = 15)
```

## Histogramas de 'Peso' y 'Cintura'

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1) #1 columna, 2 renglones, posicion 1
df['peso'].plot(kind = 'hist', figsize = (20,10))
plt.title('Histograma de Peso', fontsize = 15)


ax2 = fig.add_subplot(1, 2, 2) #1 columna, 2 renglones, posicion 2
df['cintura'].plot(kind = 'hist', figsize = (20,10))
plt.title('Histograma de Cintura', fontsize = 15)
```

## Dispersión de Peso y Cintura

```python
plt.figure(figsize = (10,5))
sns.set(font_scale = 1)

reg_plot = sns.lmplot(x = 'diastol',
            y = 'peso',
            hue = 'sexo',
            data = df)

plt.show()
```

```python
sns.boxplot(data = [df.loc[df['sexo'] == 2,
                      'cintura']]).set_title('Box plot de cintura de sexo 2')
```

```python
plt.figure(figsize = (15,10))
sns.set(font_scale = 1.7)

reg_plot = sns.regplot(x = df.loc[df['sexo'] == 1,
                      'sistol'],
            y = df.loc[df['sexo'] == 1,
                      'diastol'],
            data = df,
            color = 'red',
            marker = '+')

reg_plot.set_title('Regresión de Hombre_Sistol y Hombre_Diastol', fontsize = 15)

plt.show()
```

```python
df.loc[df['sexo'] == 1,'peso'].plot(kind = 'hist', figsize = (20,10), bins = 21)
plt.title('Histograma de Hombre_Peso', fontsize = 15)
```
